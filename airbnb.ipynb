```python
#"Outliers" im Ãœbernachtungspreis pro Stadt inkl. GÃ¤stezufriedenheit

def finde_ausreisser_pro_stadt(df, spalte='realSum', faktor=1.5):

    def berechne_grenzen(serie):
        Q1 = serie.quantile(0.25)
        Q3 = serie.quantile(0.75)
        IQR = Q3 - Q1
        untere = Q1 - faktor * IQR
        obere = Q3 + faktor * IQR
        return pd.Series([untere, obere], index=['untere_grenze', 'obere_grenze'])

    grenzen = df.groupby('stadt')[spalte].apply(berechne_grenzen).unstack()

    df_mit_grenzen = df.merge(
        grenzen.reset_index(),
        on='stadt',
        how='left'
    )

    ausreisser_pro_stadt = df_mit_grenzen[
        (df_mit_grenzen[spalte] < df_mit_grenzen['untere_grenze']) | 
        (df_mit_grenzen[spalte] > df_mit_grenzen['obere_grenze'])
    ]
    
    return ausreisser_pro_stadt.drop(columns=['untere_grenze', 'obere_grenze'])

ausreisser_final = finde_ausreisser_pro_stadt(df_all, spalte='realSum')

print("--- Outliers (?) im Ãœbernachtungspreis pro Stadt ---")
print(f"Gesamtanzahl Outliers: {len(ausreisser_final)}")

print(ausreisser_final[[
    'stadt', 
    'room_type', 
    'realSum',
    'guest_satisfaction_overall'  
]].sort_values(by='realSum', ascending=False).head(10))
```

```python
from matplotlib.ticker import FuncFormatter

df_sum_zeit = df_all.groupby(['stadt', 'zeit'])['realSum'].sum().reset_index(name='Gesamtkosten')

gesamt_kosten_pro_stadt = df_sum_zeit.groupby('stadt')['Gesamtkosten'].sum()
sort_order_stadt = gesamt_kosten_pro_stadt.sort_values(ascending=False).index

df_sum_zeit['stadt'] = pd.Categorical(df_sum_zeit['stadt'], categories=sort_order_stadt, ordered=True)

def currency_formatter(x, pos):
    """Formatiert die Achsenwerte in Tsd. â‚¬ oder Mio. â‚¬."""
    if x >= 1e6:
        return f'{x*1e-6:1.1f} Mio. â‚¬'
    elif x >= 1e3:
        return f'{x*1e-3:1.0f} Tsd. â‚¬'
    return f'{x:1.0f} â‚¬' 

plt.figure(figsize=(12, 6))

ax = sns.barplot(
    data=df_sum_zeit, 
    x='stadt',
    y='Gesamtkosten',
    hue='zeit', 
    order=sort_order_stadt, 
    palette={'weekdays': "#d90c67", 'weekends': "#646364"} 
)


formatter = FuncFormatter(currency_formatter)
ax.yaxis.set_major_formatter(formatter)

ax.set_facecolor("#AFABAF") 

plt.title('Angebotskosten (Gesamtsumme) nach Stadt und Wochentyp', fontsize=16)
plt.xlabel('Stadt')
plt.ylabel('Angebotskosten (Summe)')
plt.xticks(rotation=45, ha='right')
plt.legend(title='Wochentyp')
plt.tight_layout()
plt.show()
```
